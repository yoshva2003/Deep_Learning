# BASIC LINEAR ALGEBRA EXERCISES IN PYTHON
### Week 1: Exercise 1 - Matrix Multiplication 

```python
import numpy as np
```

```python
a=np.array([[1,2],[3,4]])
b=np.array([[5,6],[7,8]])
c=np.dot(a,b)
print("Matrix of A:")
print(a)
print("Matrix of B:")
print(b)
print("Multiplication(A,B):")
print(c)
```

### Excercise:2-Eigen Decomposition

```python
A=np.array([[4,2],[1,3]])
eigenvalues,eigenvectors=np.linalg.eig(A)
print(A)
print(eigenvalues)
print(eigenvectors)
```

### Exercise 3- Singular Value Decomposition (SVD) 

```python
A=np.array([[3,2,2],[2,3,-2]])
U,S,V=np.linalg.svd(A)
print(A)
print(U)
print(S)
print(V)
```

### Exercise 4- Determinant Calculation

```python
 A = np.array([[6, 1], [2, 3]]) 
 det = np.linalg.det(A) 
 print("Matrix A:") 
 print(A) 
 print("Determinant of A:") 
 print(det) 
```

### Exercise 5 - Norm Calculation

```python
V=np.array([3,4])
l1_norm=np.linalg.norm(V,ord=1)
l2_norm=np.linalg.norm(V)
print(V)
print(l1_norm)
print(l2_norm)
```
