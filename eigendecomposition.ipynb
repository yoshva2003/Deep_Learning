```python
import numpy as np
A=np.array([[4,2],[1,3]])
eigenvalues,eigenvectors=np.linalg.eig(A)
print("Matrix of A:")
print(A)
print("Eigenvalues:")
print(eigenvalues)
print("Eigenvectors:")
print(eigenvectors)
```
