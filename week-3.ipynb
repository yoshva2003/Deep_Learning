# Singular value Decomposition

```python
import numpy as np 
a=np.array([[3,2,2],[2,3,-2]])
U,S,V=np.linalg.svd(a)
print(a)
print(U)
print(S)
print(V)
```

```python
import numpy as np 
a=np.array([[3,2],[2,-2]])
U,S,V=np.linalg.svd(a)
print(a)
print(U)
print(S)
print(V)
```

# Determinant Calculation

```python
import numpy as np
A=np.array([[6,1],[2,3]])
det=np.linalg.det(A)
print(A)
print(det)
```

# Norm Calculation

```python
import numpy as np
V=np.array([3,4])
L1_norm=np.linalg.norm(V,ord=1)
L2_norm=np.linalg.norm(V)
print(V)
print(L1_norm)
print(L2_norm)
```

```python
import numpy as np
V=np.array([3,4,5])
L1_norm=np.linalg.norm(V,ord=1)
L2_norm=np.linalg.norm(V)
print(V)
print(L1_norm)
print(L2_norm)
```
